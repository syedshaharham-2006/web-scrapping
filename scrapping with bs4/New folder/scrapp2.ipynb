```python
from bs4 import BeautifulSoup
import requests

url = 'https://saamaan.pk/collections/accessories-and-gadgets'
content = requests.get(url)
soup = BeautifulSoup(content.content, 'html.parser')

divs = soup.find_all('div', 'product-item__info')

for div in divs:
    vendors = soup.find('a', 'product-item__vendor link')
    
    if vendors:
        vendors.decompose()
    
    add_to_cart =soup.find('button', 'product-item__action-button product-item__action-button--list-view-only button button--small button--primary') 
    
    if add_to_cart:
        add_to_cart.decompose()
    
    quick_view = div.find('button', class_='product-item__action-button product-item__action-button--list-view-only button button--small button--ternary hidden-phone')
    if quick_view:
        quick_view.decompose()
    
    product_info = div.get_text(separator='\n', strip=True)


    print(product_info,"\n" )
    
```
